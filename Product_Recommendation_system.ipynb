```python
amazon_df['keywords'] = amazon_df['keywords'].str.replace(r'[^\w\d\s]', '', regex=True)
```

```python
amazon_df['keywords']
```

```python
stemmer = PorterStemmer()
```

```python
## create a function for stemming
def tokenize_stem(text):
  words = []
  for word in text.split():
    if word not in stopwords.words('english'):
      words.append(stemmer.stem(word))
  return ' '.join(words)
```

```python
amazon_df['keywords'] = amazon_df['keywords'].apply(tokenize_stem)
```

```python
amazon_df.head(2)
```

```python
## Vectorize and Cosine similarity
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Tf-Idf Feature Extraction
TfidfVectorizer = TfidfVectorizer(max_features=5000, stop_words='english', ngram_range=(1,3),dtype=np.float32)
tfidf_matrix = TfidfVectorizer.fit_transform(amazon_df['keywords']).toarray()
```

```python
# Calculate similarity
similarity = cosine_similarity(tfidf_matrix)
del(tfidf_matrix)
```

```python
amazon_df.drop(['keywords'], axis=1, inplace=True)
```

```python
def product_recommender(query):
    # Normalize case for matching
    matches = amazon_df[amazon_df['name'].str.lower() == query.lower()]
    # matches = amazon_df[amazon_df['name'].str.contains(query, case=False, na=False)]

    # Handle missing product gracefully
    if matches.empty:
        print(f"❌ No product found matching '{query}'. Try a different name or partial match.")
        return

    # Get the first matching index
    product_index = matches.index[0]

    # Get cosine similarity scores
    similarity_list = list(enumerate(similarity[product_index]))

    # Sort by similarity score (highest first)
    top_10_similar_products = sorted(similarity_list, key=lambda x: x[1], reverse=True)[1:11]

    # Display top similar products
    print(f"\nTop recommendations for '{query}':\n")
    for idx, similarity_score in top_10_similar_products:
        print(f"- {amazon_df.iloc[idx]['name']}")
```

```python
## Reommendation System
product_recommender('MI Poco M2 (Brick Red, 6GB RAM, 128GB Storage)')
```
